## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

##
This is a ranking/scoring problem, not classification. The decision editors face isn't "is this
page declining, yes or no" -- it's "given limited weekly review capacity, which pages do I look
at first." That's a "which ones first?" question, which maps to ranking/scoring, not a binary
label consumed on its own. The output is a priority score used to sort a queue, not a
standalone yes/no verdict.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed if-statement rule already exists here as the baseline (the starter pipeline's rule
baseline), and it gets precision@50 = 0.24 -- meaning roughly 3 out of 4 pages it flags aren't
actually worth an editor's time. The learned model on the same data gets precision@50 = 0.74.
That ~3x gap is the evidence, not a guess: the signals that separate a real opportunity from
noise (CTR, scroll_rate, avg_position, word_count, content_type-dependent missingness) are individually
weak and interact with each other, so no single threshold on one or two columns captures the
pattern without either flooding editors with false positives or missing real declines.
